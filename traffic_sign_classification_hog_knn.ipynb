## Computing HoG Features

```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt
```

```python
features = []
hog_images = []
for i in train:
  image = i
  resized_img = resize(image, (128*4, 64*4))
  fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8),
                	cells_per_block=(2, 2), visualize=True, multichannel=True)
  features.append(fd)
  hog_images.append(hog_image)
  # plt.axis("off")
  # plt.imshow(resized_img)
  # print(resized_img.shape)
```

## K-Means Clustering

```python
''''kmeans'''

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

n_clusters = 10
n_samples = 1500
epochs = 500
features, l_trn = make_blobs(n_samples=n_samples, centers=n_clusters, cluster_std=0.60, random_state=0)

kmeans = KMeans(n_clusters=n_clusters, random_state=0, max_iter=epochs).fit(features)    

fig, ax = plt.subplots(figsize=(10, 6))
ax1 = plt.subplot(1, 1, 1)
ax1.scatter(features[:, 0], features[:, 1], c='tab:blue', s=25)
for k in kmeans.cluster_centers_:
    ax1.scatter(k[0], k[1], c='red', s=50)
ax1.set_title("Scikit-Learn K-means Results")
plt.show()
```

```python
n_clusters = 10
n_samples = 1500
epochs = 500
tst_features, l_tst = make_blobs(n_samples=n_samples, centers=10, cluster_std=0.60, random_state=0)
predict_kmeans = kmeans.predict(tst_features)

fig, ax = plt.subplots(figsize=(10, 6))
ax1 = plt.subplot(1, 1, 1)
ax1.scatter(tst_features[:, 0], tst_features[:, 1], c='tab:blue', s=25)
for k in kmeans.cluster_centers_:
    ax1.scatter(k[0], k[1], c='red', s=50)
ax1.set_title("Scikit-Learn K-means Results")
plt.show()
```
